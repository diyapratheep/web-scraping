```python
from bs4 import BeautifulSoup
import requests
```

```python
url ='https://en.wikipedia.org/wiki/List_of_prime_ministers_of_India'
page = requests.get(url) # uses the requests library to send a GET request to the URL and retrieve the webpage
soup = BeautifulSoup(page.text,'html') #to parse the HTML content of the webpage
```

```python
print(soup)
```

```python
table = soup.find_all('table')[1]
table
```

```python
soup.find('table', class_ = "wikitable sortable sticky-header-multi sort-under") #oh, its same same, hehe
```

```python
table_titles = table.find_all('th') #extracting columns only
table_titles
```

```python
list_of_titles = [titles.text.strip() for titles in table_titles][:9]
list_of_titles
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns=list_of_titles)
df
```

```python
rows = table.find_all('tr')
rows
```

data_list =[]
for i in rows[2:]:
    row_data = i.find_all('td')
    table_data = [x.text.strip() for x in row_data]
    #df.loc[len(df)] = table_data
    data_list.append(table_data)

df = pd.DataFrame(data_list[:9], columns=list_of_titles)
